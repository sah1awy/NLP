```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.metrics import adjusted_rand_score
import numpy as np
import spacy
```

```python
docs = ["This little kitty came to play when I was eating at a restaurant.",
             "Merley has the best squooshy kitten belly.",
             "Google Translate app is incredible.",
             "If you open 100 tab in google you get a smiley face.",
             "Best cat photo I've ever taken.",
             "Climbing ninja cat.",
             "Impressed with google map feedback.",
             "Key promoter extension for Google Chrome."]
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.Defaults.stop_words.remove("not")
nlp.vocab["not"].is_stop = False
```

```python
x = np.array(docs)
def process(doc):
    doc = nlp(str(doc))
    l = []
    for tok in doc:
        if not (tok.is_stop or tok.is_punct or tok.is_space):
            l.append(tok.lemma_)
    return " ".join(l)
```

```python
vfunc = np.vectorize(process)
txt = vfunc(x)
```

```python
tf_idf = TfidfVectorizer()
txt_vec = tf_idf.fit_transform(txt)
```

```python
km = KMeans(n_clusters = 2,init='k-means++')
km.fit(txt_vec)
```

```python
print("Top terms per cluster:")
```

```python
cen = km.cluster_centers_.argsort()[:,::-1]
cen
```

```python
terms = tf_idf.get_feature_names_out()
```

```python
for c in range(2):
    print(f"Cluster: {c}")
    for x in cen[c,:10]:
        print(" %s"%terms[x])
    print('-'*50)
```

```python
y = tf_idf.transform(["chrome browser to open."])
prediction = km.predict(y)
print(prediction)
```
